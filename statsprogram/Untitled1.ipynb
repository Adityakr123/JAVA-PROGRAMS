```python
import pandas as pd
import numpy as np
import statistics as st
import matplotlib.pyplot as plt

df = pd.read_excel(r'Demo.xlsx')
arrmarks=[]
arrcgpi=[]
covariance=0
for i in range (len(df)):
    arrmarks.append(df['Marks'][i])
    arrcgpi.append(df['CGPI'][i])
plt.plot(arrmarks, arrcgpi)
plt.xlabel('marks')
plt.ylabel('cgpi')
plt.show()

for i in range(len(arrmarks)):
    covariance=covariance+(((arrmarks[i]-st.mean(arrmarks))*(arrcgpi[i]-st.mean(arrcgpi)))/len(arrmarks))
x=st.stdev(arrmarks)*st.stdev(arrcgpi)
x=covariance/x
print("value of covariance  is:-",covariance)
print("coefficent of corelation  is:-",x)
```
